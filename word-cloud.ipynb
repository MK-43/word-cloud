```python
from wordcloud import WordCloud
from konlpy.tag import Okt
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Sample Korean text
with open('text.txt', 'r', encoding='utf-8') as f:
    text = f.read()

# Define Korean stop words
with open('stopwords.txt', 'r', encoding='utf-8') as f:
    stop_words = f.read()

stop_words = set(stop_words)
print(stop_words)
```

```python
# 1. Tokenize and extract nouns
okt = Okt()
nouns = okt.nouns(text)

# 2. Remove stop words
filtered_nouns = [word for word in nouns if word not in stop_words and len(word) > 1]

# 3. Count word frequencies
word_count = Counter(filtered_nouns)

# 4. Generate the word cloud
wordcloud = WordCloud(
    font_path='NanumGothic.ttf',  # Change to your Korean font path
    background_color='white',
    width=800,
    height=600
).generate_from_frequencies(word_count)

# 5. Show the word cloud
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Korean Word Cloud (With Stop Words)", fontsize=20)

plt.savefig('wordcloud.png', bbox_inches='tight', dpi=300)

plt.show()
```
